```python
import matplotlib.pyplot as plt
# 读取
data = []
with open('results.txt', 'r') as file:
    for line in file:
        parts = line.split()
        loop = int(parts[0])
        loop_type = parts[1]
        sumtime = float(parts[-1])
        data.append((loop, loop_type, sumtime))

# 数据按循环类型分组
loop1_data = [(loop, sumtime) for loop, loop_type, sumtime in data if loop_type == 'loop1']
loop2_data = [(loop, sumtime) for loop, loop_type, sumtime in data if loop_type == 'loop2']
loop3_data = [(loop, sumtime) for loop, loop_type, sumtime in data if loop_type == 'loop3']

# 绘制图表
plt.plot(*zip(*loop1_data), label='loop1')
plt.plot(*zip(*loop2_data), label='loop2')
plt.plot(*zip(*loop3_data), label='loop3')

# 设置图表标题和标签
plt.title('3 loops contrast')
plt.xlabel('repeat times')
plt.ylabel('SUM time')

plt.legend()

plt.show()
```

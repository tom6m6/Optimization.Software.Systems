任务5绘图

```python
import matplotlib.pyplot as plt

with open('results.txt', 'r') as file:
    lines = file.readlines()

data = {}
current_compiler = None

for line in lines:
    line = line.strip()
    if line == 'gcc' or line == 'clang':
        current_compiler = line
        data[current_compiler] = {'inputs': [], 'runtimes': []}
    else:
        input_val, runtime = map(float, line.split(' : '))
        data[current_compiler]['inputs'].append(input_val)
        data[current_compiler]['runtimes'].append(runtime)

plt.plot(data['gcc']['inputs'], data['gcc']['runtimes'], label='gcc')
plt.plot(data['clang']['inputs'], data['clang']['runtimes'], label='clang')

plt.title('Time with input')
plt.xlabel('input')
plt.ylabel('Time/sec')

# 图例
plt.legend()

# 在每个数据点旁边显示数值
for compiler, values in data.items():
    for x, y in zip(values['inputs'], values['runtimes']):
        plt.text(x, y, f'{y:.3f}', ha='left', va='bottom', fontsize=10, color='black')

plt.show()
```

```python
import matplotlib.pyplot as plt

with open('results_merge.txt', 'r') as file:
    lines = file.readlines()

data = {}
current_compiler = None

for line in lines:
    line = line.strip()
    if line.startswith('gcc') or line.startswith('clang'):
        current_compiler = line
        data[current_compiler] = {'inputs': [], 'runtimes': []}
    else:
        input_val, runtime = map(float, line.split(' : '))
        data[current_compiler]['inputs'].append(input_val)
        data[current_compiler]['runtimes'].append(runtime)

plt.plot(data['gcc-qemu-aarch64']['inputs'], data['gcc-qemu-aarch64']['runtimes'], label='gcc-qemu-aarch64')
plt.plot(data['clang-qemu-aarch64']['inputs'], data['clang-qemu-aarch64']['runtimes'], label='clang-qemu-aarch64')
plt.plot(data['gcc-host']['inputs'], data['gcc-host']['runtimes'], label='gcc-host')
plt.plot(data['clang-host']['inputs'], data['clang-host']['runtimes'], label='clang-host')

plt.title('Time vs Input')
plt.xlabel('Input')
plt.ylabel('Time/sec')

# 图例
plt.legend()

# 在每个数据点旁边显示数据
for compiler, values in data.items():
    for x, y in zip(values['inputs'], values['runtimes']):
        plt.text(x, y, f'{y:.2f}', ha='left', va='bottom', fontsize=8, color='black')

plt.show()
```
